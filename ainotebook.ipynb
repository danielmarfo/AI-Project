```python
from pptx import Presentation

def read_pptx_file(file_path):
    """
    Function to read text content from a PowerPoint (.pptx) file.

    Parameters:
    file_path (str): The file path to the PowerPoint (.pptx) file.

    Returns:
    text_content (str): The concatenated text content extracted from all slides.
    """
    # Initialize an empty string to store the extracted text content
    text_content = ""

    # Load the PowerPoint presentation
    presentation = Presentation(file_path)

    # Iterate through each slide in the presentation
    for slide in presentation.slides:
        # Iterate through each shape in the slide
        for shape in slide.shapes:
            # Check if the shape has a text frame
            if hasattr(shape, "text"):
                # Concatenate the text content of the shape to the result
                text_content += shape.text + "\n"

    return text_content

# Example usage:
file_path = "lecture.pptx"
text_content = read_pptx_file(file_path)
print(text_content)
```

```python
import PyPDF2

def read_pdf_file(file_path):
    """
    Function to read text content from a PDF file.

    Parameters:
    file_path (str): The file path to the PDF file.

    Returns:
    text_content (str): The text content extracted from the PDF file.
    """
    # Open the PDF file
    with open(file_path, "rb") as file:
        # Initialize a PDF file reader object
        pdf_reader = PyPDF2.PdfFileReader(file)
        
        # Initialize an empty string to store the extracted text content
        text_content = ""
        
        # Iterate through each page in the PDF file
        for page_num in range(pdf_reader.numPages):
            # Extract text from the page
            page = pdf_reader.getPage(page_num)
            text_content += page.extractText()

    return text_content
gvjbjbjhbhjbibhkblkh
# Example usage:
file_path = "lecture1.pdf"
text_content = read_pdf_file(file_path)
print(text_content)
```
